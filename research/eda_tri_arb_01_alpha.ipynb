```python
import polars as pl
from polars import col
pl.Config.set_tbl_rows(100)
```

```python
quote_pattern = r"(USDT|BUSD|USDC|EUR|USD|BTC|ETH)$"
```

```python
df = pl.read_ndjson("../datalake/ingest_20260610_181650.jsonl").slice(1)\
    .with_columns(col("log_ts_utc_ms").cast(pl.Datetime("ms")).alias("datetime"))\
    .select('datetime', 'symbol', 'last_update_id', 'best_bid', 'best_bid_qty', 'best_ask', 'best_ask_qty')
   
    #.with_columns([ col("symbol").str.extract(quote_pattern, 1).alias("qoute"), col("symbol").str.replace(quote_pattern, "").alias("base")])
```
